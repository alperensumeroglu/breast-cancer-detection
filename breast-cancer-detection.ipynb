```python
diagnosis_counts = data['diagnosis'].value_counts()
print(diagnosis_counts)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5,2))
plt.bar(diagnosis_counts.index, diagnosis_counts.values, color = ["red", "blue"], alpha=0.7)
plt.title("Patient Diagnosis Distribution")
plt.xlabel("Diagnosis")
plt.ylabel("Number of patients")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
data['diagnosis_numeric'] = data['diagnosis'].map({'M': 1, 'B': 0})

plt.figure(figsize=(5, 2))  
plt.scatter(data['id'], data['radius_mean'], s=data['diagnosis_numeric']*20)
plt.xlabel('ID')  
plt.ylabel('Radius Mean') 
plt.title('Bubble Chart of Diagnosis and Radius Mean')  
plt.grid(True)  
plt.show()
plt.accoording--
```
